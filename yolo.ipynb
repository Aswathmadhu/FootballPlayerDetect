```python
import ultralytics
import os
import shutil
from tqdm.notebook import tqdm
```

```python
import random
random.seed(42)
```

```python
# train_img_path = '/content/yolo_new/images/train'
train_img_path = './yolo_new/images/train'
train_label_path = './yolo_new/labels/train'
val_img_path = './yolo_new/images/val'
val_label_path = './yolo_new/labels/val'
```

```python
# defining function
def train_test_split(path,split=0.2):
    files = list(set([name[:-4] for name in os.listdir(path)]))
    #[:-4]=> to remove extention and set used to remove duplicates then saved to list
    # print("names of files :",files)
    # print("number of files :",len(files))
    random.shuffle(files)

    # test size
    test_size = int(len(files)*split)
    print("test size :",test_size)

    # trian size
    train_size = len(files)-test_size
    print("train size :",train_size)

    # create folders
    os.makedirs(train_img_path,exist_ok=True)
    os.makedirs(train_label_path,exist_ok=True)
    os.makedirs(val_img_path,exist_ok=True)
    os.makedirs(val_label_path,exist_ok=True)

    # copying into files using shutil
    # shutil.copy2 source,destination

    for i in tqdm(files[:train_size]):
        shutil.copy2(os.path.join(path,i+'.jpg'),f"{train_img_path}/"+i+".jpg")
        shutil.copy2(os.path.join(path,i+'.txt'),f"{train_label_path}/"+i+".txt")
    for i in tqdm(files[:test_size]):
        shutil.copy2(os.path.join(path,i+'.jpg'),f"{val_img_path}/"+i+".jpg")
        shutil.copy2(os.path.join(path,i+'.txt'),f"{val_label_path}/"+i+".jpg")

    
```
